```
import matplotlib.pyplot as plt
```

```
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y,'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax.set_ylabel('Z')

ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
```

```
fig,axes = plt.subplots(1,2) #nrows=1,ncols=2
```

```
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
plt.tight_layout()
fig
```

```
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
axes[0].plot(x,y,color="blue", lw=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
